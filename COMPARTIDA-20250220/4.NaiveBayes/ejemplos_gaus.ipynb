üìå 1. Na√Øve Bayes Gaussiano
Se usa cuando las caracter√≠sticas siguen una distribuci√≥n normal (gaussiana). Es com√∫n en clasificaci√≥n con datos continuos.

üîπ Ejemplo: Clasificaci√≥n de flores Iris

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Cargar datos de Iris
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Modelo Na√Øve Bayes Gaussiano
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Predicciones
y_pred = gnb.predict(X_test)

# Evaluaci√≥n
print(f"Precisi√≥n: {accuracy_score(y_test, y_pred):.2f}")
```

‚úÖ Uso t√≠pico: Clasificaci√≥n de datos continuos como an√°lisis de c√°ncer o reconocimiento de patrones.


üìå 2. Na√Øve Bayes Multinomial
Se usa cuando los datos representan conteos de eventos (por ejemplo, frecuencia de palabras en texto).

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Datos de ejemplo
emails = ["Gana dinero r√°pido", "Oferta exclusiva para ti", "Reuni√≥n ma√±ana a las 10", "Descuento en tu pr√≥xima compra"]
labels = [1, 1, 0, 1]  # 1 = spam, 0 = no spam

# Convertir textos a frecuencias
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.5, random_state=42)

# Modelo Na√Øve Bayes Multinomial
mnb = MultinomialNB()
mnb.fit(X_train, y_train)

# Predicciones
y_pred = mnb.predict(X_test)

# Evaluaci√≥n
print(f"Precisi√≥n: {accuracy_score(y_test, y_pred):.2f}")
```

‚úÖ Uso t√≠pico: Clasificaci√≥n de texto, an√°lisis de sentimiento, detecci√≥n de spam.

üìå 3. Na√Øve Bayes Bernoulli
Se usa cuando los datos son binarios (presencia o ausencia de una caracter√≠stica)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score

# Datos de ejemplo
emails = ["Compra ahora", "Descuento especial", "Hola, ¬øc√≥mo est√°s?", "Oferta limitada"]
labels = [1, 1, 0, 1]  # 1 = spam, 0 = no spam

# Convertir textos a representaci√≥n binaria
vectorizer = CountVectorizer(binary=True)
X = vectorizer.fit_transform(emails)

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.5, random_state=42)

# Modelo Na√Øve Bayes Bernoulli
bnb = BernoulliNB()
bnb.fit(X_train, y_train)

# Predicciones
y_pred = bnb.predict(X_test)

# Evaluaci√≥n
print(f"Precisi√≥n: {accuracy_score(y_test, y_pred):.2f}")
```

‚úÖ Uso t√≠pico: Clasificaci√≥n de documentos con presencia o ausencia de palabras clave (detecci√≥n de spam con palabras importantes).

Gaussian Na√Øve Bayes (GaussianNB): se trata de una variante del clasificador Na√Øve Bayes, que se utiliza con distribuciones gaussianas, es decir, distribuciones normales y variables continuas. Este modelo se ajusta hallando la media y la desviaci√≥n t√≠pica de cada clase.  

Multinomial Na√Øve Bayes (MultinomialNB): este tipo de clasificador Na√Øve Bayes asume que las caracter√≠sticas provienen de distribuciones multinomiales. Esta variante es √∫til cuando se utilizan datos discretos, como recuentos de frecuencia, y suele aplicarse en casos de uso de procesamiento del lenguaje natural, como la clasificaci√≥n del spam.  

Bernoulli Na√Øve Bayes (BernoulliNB): se trata de otra variante del clasificador Na√Øve Bayes, que se utiliza con variables booleanas, es decir, variables con dos valores, como verdadero y falso o 1 y 0.  


## üîç ¬øCu√°l modelo usar?

| Modelo         | Datos                                                    |
|---------------|----------------------------------------------------------|
| **Gaussiano**  | Variables continuas (ej. reconocimiento de d√≠gitos, datos biom√©tricos) |
| **Multinomial** | Conteo de palabras en documentos (ej. NLP, detecci√≥n de spam) |
| **Bernoulli**  | Datos binarios (ej. detecci√≥n de spam con presencia/ausencia de palabras clave) |


Aqu√≠ tienes una explicaci√≥n sencilla de cu√°ndo usar cada variante de Na√Øve Bayes:  

üü¢ Gaussiano (GaussianNB)  
üìå ¬øCu√°ndo usarlo?  
Cuando las caracter√≠sticas de los datos son n√∫meros continuos (como altura, peso, temperatura, etc.).  
üìä Ejemplo real: Diagn√≥stico de enfermedades basado en an√°lisis de sangre.  

üîµ Multinomial (MultinomialNB)  
üìå ¬øCu√°ndo usarlo?  
Cuando los datos representan conteos de eventos, como la frecuencia de palabras en textos.  
üìä Ejemplo real: Clasificaci√≥n de correos spam usando la cantidad de veces que aparecen ciertas palabras.  

üî¥ Bernoulli (BernoulliNB)  
üìå ¬øCu√°ndo usarlo?  
Cuando los datos son binarios (s√≠/no, 1/0, verdadero/falso), es decir, cuando solo importa la presencia o ausencia de algo.  
üìä Ejemplo real: Detecci√≥n de spam basada en si ciertas palabras aparecen o no en el correo.  

üîç ¬øC√≥mo elegir?  
Datos num√©ricos continuos? ‚Üí Usa Gaussiano.  
Datos basados en conteos (como textos)? ‚Üí Usa Multinomial.  
Datos binarios (presente/ausente)? ‚Üí Usa Bernoulli.  
üöÄ Consejo: Para clasificaci√≥n de texto, MultinomialNB es el m√°s com√∫n.
