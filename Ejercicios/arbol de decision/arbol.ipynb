### dame un código en python para implementar arbol de decisicion binario donde el usuario responda a preguntas de tipo si o no, para conocer si se trata de una especie de arbol o de otra., por ejemplo: pino,cipres,fresno, cedro

```python
import json
import os

def cargar_o_crear_preguntas():
    nombre_archivo = "preguntas_arboles.json"
    if os.path.exists(nombre_archivo):
        with open(nombre_archivo, 'r', encoding='utf-8') as archivo:
            return json.load(archivo)
    else:
        preguntas = {
            "aciculares": {
                "pregunta": "¿El árbol tiene hojas en forma de aguja?",
                "si": "tamaño_aciculas",
                "no": "no_aciculares"
            },
            "tamaño_aciculas": {
                "pregunta": "¿El árbol tiene las agujas largas?",
                "si": "Pino",
                "no": "aciculas_cortas"
            },
            "aciculas_cortas": {
                "pregunta": "¿El árbol tiene las agujas reunidas en grupos?",
                "si": "Cedro",
                "no": "Abeto"
            },


            "no_aciculares": {
                "pregunta": "¿Las hojas tiene el limbo plano y ancho?",
                "si": "tipo_de_hojas",
                "no": "Ciprés"
            },

             "tipo_de_hojas": {
                "pregunta": "¿Las hojas son compuestas?",
                "si": "tipo_de_hojas_compuesta",
                "no": "tipo_hojas_simples"
            },

            "tipo_de_hojas_compuesta": {
            "pregunta": "¿Las hojas son palmadao-compuestas?",
            "si": "Castaño de indias",
            "no": "hojas_pinnado_compuestas_fruto"
            },

            "hojas_pinnado_compuestas_fruto": {
            "pregunta": "¿El fruto es una legumbre?",
            "si": "Falsa acacia",
            "no": "Fresno"
            },

            "tipo_hojas_simples": {
            "pregunta": "¿Las hojas son duras?",
            "si": "hojas_duras_fruto",
            "no": "tipo_hojas_blandas"
            },

         "hojas_duras_fruto": {
            "pregunta": "¿El fruto es una bellota?",
            "si": "Encina",
            "no": "Madroño"
            },

       
        "tipo_hojas_blandas": {
        "pregunta": "¿La hoja es Penninervia?",
        "si": "tipo_hoja_penninervia",
        "no": "tipo_hoja_palminervia"
        },

        "tipo_hoja_penninervia": {
        "pregunta": "¿La hoja es 2 veces más larga que ancha?",
        "si": "Suace llorón",
        "no": "tipo_hoja_penninervia_borde"
        },

        "tipo_hoja_penninervia_borde": {
        "pregunta": "¿La hoja tiene el borde inferior asimétrico, aserrado y ovalado?",
        "si": "Olmo",
        "no": "Chopo"
        },



            "Suace llorón":"Suace llorón",
            "Pino": "Pino",
            "Olmo":"Olmo",
            "Chopo":"Chopo",
            "Encina": "Encina",
            "Madroño":"Madroño",
            "Ciprés": "Ciprés",
            "Cedro": "Cedro",
            "Fresno": "Fresno",
            "Abeto": "Abeto",
            "Castaño de indias":"Castaño de indias",
            "Falsa acacia":"Falsa acacia",
            "Desconocido": "Desconocido"
        }
        with open(nombre_archivo, 'w', encoding='utf-8') as archivo:
            json.dump(preguntas, archivo, ensure_ascii=False, indent=4)
        return preguntas

def arbol_decision(preguntas):
    print("Bienvenido al identificador de árboles. Responde 'si' o 'no' a las siguientes preguntas.")
    
    nodo_actual = "aciculares"
    while True:
        if not isinstance(preguntas[nodo_actual], dict):  # Si llegamos a una hoja (especie de árbol)
            return preguntas[nodo_actual]
        
        respuesta = pregunta(preguntas[nodo_actual]["pregunta"])
        nodo_actual = preguntas[nodo_actual]["si"] if respuesta else preguntas[nodo_actual]["no"]

def pregunta(texto):
    while True:
        respuesta = input(texto + " (si/no): ").lower()
        if respuesta in ["si", "no"]:
            return respuesta == "si"
        else:
            print("Por favor, responde 'si' o 'no'.")

# Cargar o crear el archivo de preguntas
preguntas = cargar_o_crear_preguntas()

# Ejecutar el árbol de decisión
especie = arbol_decision(preguntas)
print(f"El árbol identificado es: {especie}")
```
